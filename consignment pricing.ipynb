```python
df=df[['Managed By','Fulfill Via','Shipment Mode','Weight (Kilograms)','cons_price']]
```

```python
# From the code no. 37 we can say that most of the logistics is managed by the PMO, So we can drop the column 
```

```python
df = pd.get_dummies(df,columns=["Managed By","Fulfill Via","Shipment Mode"])
```

```python
df.info()
```

```python
sns.pairplot(df)
```

```python
X = df.drop("cons_price",axis=1) 
y= df["cons_price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
y_pred
```

```python
y_mse = mean_squared_error(y_test,y_pred)
y_rmse = np.sqrt(y_mse)
y_rmse
```

```python
from sklearn.metrics import r2_score
r2l = r2_score(y_test,y_pred)
print(r2l)
```

```python
n=40
k=2
adj_r2_score = 1 - ((1-r2l)*(n-1)/(n-k-1))
print(adj_r2_score)
```

## Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
rg = DecisionTreeRegressor()
rg.fit(X_train,y_train)
```

```python
y_pred1 = rg.predict(X_test)
```

```python
y_pred1
```

```python
y_mse1 = mean_squared_error(y_test,y_pred1)
y_rmse1 = np.sqrt(y_mse1)
y_rmse1
```

```python
from sklearn.metrics import r2_score
r2d = r2_score(y_test,y_pred1)
print(r2d)
```

```python
## Here error is lesser than the Linear Regression, Hence Decision tree is better model than linear regression
```

```python
n=40
k=2
adj_r2_score = 1 - ((1-r2d)*(n-1)/(n-k-1))
print(adj_r2_score)
```

## Random Forest
